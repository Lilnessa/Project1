```python
# Read in CSV and clean data
SD_median_income = pd.read_csv("Resources/Household Income Median.csv")
SD_median_income = SD_median_income.rename(columns={'Zip':'zip code'})
SD_median_income.head()
```

```python
# Find unique number of zipcodes in income dataset
SD_median_income = SD_median_income.sort_values(by='zip code')
SD_median_income = SD_median_income.drop(34)
SD_median_income['zip code'].unique()
```

```python
# Find unique number of zipcodes in gas dataset
extra_clean_gas = extra_clean_gas.sort_values(by='zip code')
extra_clean_gas['zip code'].unique()
```

```python
# Calculate average gas prices per zip code
averaged_gas1_prices = extra_clean_gas.groupby(["zip code"])["gas1 price"].mean()
averaged_gas2_prices = extra_clean_gas.groupby(["zip code"])["gas2 price"].mean()
averaged_gas3_prices = extra_clean_gas.groupby(["zip code"])["gas3 price"].mean()
```

```python
# Create df with avg gas prices per zip
gas_price_per_zip_avg = pd.DataFrame(
    {
        "Regular avg": averaged_gas1_prices,
        "Midgrade avg": averaged_gas2_prices,
        "Premium avg": averaged_gas3_prices,
    }
)
gas_price_per_zip_avg.head()
gas_price_per_zip_avg.reset_index(inplace=True)
```

```python
# Load in csv with lat, long, population, density
zip_df = pd.read_csv("Resources/uszips.csv")
sd_zip_df_unclean = zip_df.loc[zip_df["county_names_all"] == "San Diego", :]
sd_zip_df_unclean = sd_zip_df_unclean.rename(columns={'zip':'zip code'})
# Create df of only info we want
sd_zip_df = sd_zip_df_unclean[['zip code', 'lat', 'lng', 'city', 'population', 'density']]
sd_zip_df = pd.DataFrame(sd_zip_df)

len(sd_zip_df)

```

```python
# Making sure zip in all dfs are string types
sd_zip_df['zip code'] = sd_zip_df['zip code'].astype(str)
gas_price_per_zip_avg['zip code'] = gas_price_per_zip_avg['zip code'].astype(str)
SD_median_income['zip code'] = SD_median_income['zip code'].astype(str)
```

```python
merg_zip_data = sd_zip_df.merge(gas_price_per_zip_avg, on=['zip code'], how='right')
merg_zip_data.head()
```
